## 1. Dataset Structure

Expected layout after downloading from Kaggle:
```
data/
    Training/
        glioma/        # ~1321 images
        meningioma/    # ~1339 images
        no_tumor/      # ~1595 images
        pituitary/     # ~1457 images
    Testing/
        ...
```

Download: https://www.kaggle.com/datasets/masoudnickparvar/brain-tumor-mri-dataset

````python
DATA_DIR = Path('../data')  # Change this to your data directory
CLASS_NAMES = ['glioma', 'meningioma', 'no_tumor', 'pituitary']

if DATA_DIR.exists():
    counts = {}
    for cls in CLASS_NAMES:
        for split in ['Training', 'Testing']:
            p = DATA_DIR / split / cls
            if p.exists():
                counts[f'{split}/{cls}'] = len(list(p.glob('*.jpg')))
    
    print('Dataset class counts:')
    for k, v in sorted(counts.items()):
        print(f'  {k}: {v} images')
else:
    print('Data directory not found. Please download the dataset first.')
    print('Then set DATA_DIR to the correct path.')
````

## 2. Sample Images per Class

````python
if DATA_DIR.exists():
    fig, axes = plt.subplots(2, 4, figsize=(14, 7))
    
    for col, cls in enumerate(CLASS_NAMES):
        for row, split in enumerate(['Training', 'Testing']):
            img_dir = DATA_DIR / split / cls
            imgs = list(img_dir.glob('*.jpg'))
            if imgs:
                img = mpimg.imread(str(imgs[0]))
                axes[row, col].imshow(img, cmap='gray')
                axes[row, col].set_title(f'{cls}\n({split})')
                axes[row, col].axis('off')
    
    plt.suptitle('Sample MRI Images per Class', fontsize=14)
    plt.tight_layout()
    plt.show()
else:
    print('Skipping — data directory not found.')
````
